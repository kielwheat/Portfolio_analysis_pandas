## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns

def balanced_weighting(): 
    weights= []
    count = 0
    weight_count = len(daily_returns.columns)
    balanced_weight = 1/weight_count
    while count < weight_count:
        weights.append(balanced_weight)
        count += 1
    return weights
        
weights = balanced_weighting()
   
my_portfolio_returns = daily_returns.dot(weights)

my_portfolio_returns
```

### How does your portfolios Sharpe ratio perform?

The custom portolio came in 2nd with a Sharpe only lower than algo1. Given that it also produced the 2nd highest returns, it is safe to say that the additional volatility was really extra reward and that is why, like Algo1,the sharpe ratio is rather low.  Generally a Sharpe above 2 is considered excellent, yet no portfolios achieved this.  The downside to the Sharpe ratio calculation is that all volatility is the same- a better ratio like a Sortino which only penalizes downside volatility or the Gain to Pain (GPR) which sums the portfolio gains divided by the absolute value of total losses.  Either of these rations provides a better idea of how well fund produces gains relative to their realized drawdowns.

```python
#User-added ratios


def sortino(all_returns):
    sortino={}
    for column in all_returns:      
        sortino[column] = round((all_returns[column].mean()*252)/(all_returns[column][all_returns[column]<0].std()*np.sqrt(252)),4)
    return sortino



#User-added Gain-to-Pain ratio
def gain_to_pain(all_returns):
    gain_to_pain={}
    for column in all_returns:
        gain_to_pain[column] = round(all_returns[column][all_returns[column]>0].sum()/abs(all_returns[column][all_returns[column]<0].sum()),4)
    return gain_to_pain



            
def all_plots():
    ratio_master = {}
    
    for key in sortino.keys():
        
        ratio_list = []
        ratio_list.append(sharpe_ratios[key])
        ratio_list.append(sortino[key])
        ratio_list.append(gpr[key])
        ratio_master[key] = ratio_list
        
    df = pd.DataFrame.from_dict(ratio_master).transpose()
    df.columns = ['Sharpe','Sortino','GPR']   
    df.plot(kind = 'bar', figsize = (20,10))  
    print(df)


sortino= sortino(all_returns)            
gpr = gain_to_pain(all_returns)
all_plots()


```

### Sharpe, Sortino and GPR analysis:
Results show some interesting highlights.
Of the porfolios that outperformed near S&P cumulative returns (Algo1, Custom, Berkshire):
Algo1 and Custom show the power of multiple ratio analysis.  The lack of highly volatile negative returns (Sharpe) and smaller drawdown to total returns (GPR) show that Algo1 and to a lesser extent, Custom, is efficient in capturing upside volatility while keeping downside risk to a minimum relative to other portfolios.
